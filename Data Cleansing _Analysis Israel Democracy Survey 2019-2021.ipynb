**Q4**

```python
# Scale dataset to have ALL features on same scale between [0-1]
from sklearn.preprocessing import MinMaxScaler # לכל מימד יש את הסקאלה שלה ואנחנו נרצה להביא את הנתונים לאותה סקאלה ולנרמל למשל לפי min max scler נייבא ספריה
scaler = MinMaxScaler() #ניייצר סקאלה בין 0 ל 1
LEFT_RIGHT_DF = pd.DataFrame(data=scaler.fit_transform(frame_copy2),columns=frame_copy2.columns) #fit לסקאלה בין 0 ל 1 הופכת הכל לטבלה כששמות העמודות לפי הטבלה מקודם תלמד מהנתונים ותהפוך אותו
LEFT_RIGHT_DF.head() # טבלה לאחר נרמול
```

**Q7**
 Silhouette

```python
from sklearn.metrics import silhouette_score #נייבא ספריה על מנת לבדוק את ממד סילואט
from sklearn.cluster import KMeans
silhouette_list = [] #ניצור רשימה
for K_iter in range(2,6): #2,3,4,5 נבדוק את כל האופציות האלה
    kmeans = KMeans(n_clusters=K_iter, init='random') #נאתחל את K לפי הפרמטר שרץ על הלולאה והאתחול הרנדומלי של הצנטרואידים
    kmeans.fit(reduced) #יוצרת טבלה דו מימדית הרצה לאימון יצרתי מודל אני רוצה להפעיל
    silhouette_list.append([K_iter,silhouette_score(reduced, kmeans.labels_)]) #להוסיף לרשימה את K ואת הסיולאט
```

```python
silhouette_list
 #נבחר ב k=3 משום שניתן לראות כי שם המדד הוא הגבוה ביותר, ייתן את החלוקה הטובה ביותר
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='random')
kmeans.fit(reduced)
```

```python
centroids_df = pd.DataFrame(data=kmeans.cluster_centers_,columns=reduced.columns)
```

```python
reduced['Predicted Class'] = kmeans.predict(reduced) #נרצה לראות את kmeans קורה באמצעות גרף ורנאה את החלוקה
```

```python
from sklearn import datasets # נבחר בגרף שכל הקבוצות עוברות את הקו האדום וגם הקבוצות הכי הומוגניות נעשה זאת בנוסף לסיולאט על מנת לבחור את הקליסתור הטוב כל הקוד הזה מייצר את הגרפים האלה
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer


fig, ax = plt.subplots(3, 2, figsize=(15,8))
for i in [2, 3, 4, 5,6,7]:
    '''
    Create KMeans instances for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(reduced)
```

**Q7**
Elbow

```python
from sklearn.cluster import KMeans
K_list = [] #רשימה של k
WCSS_list = [] #רשימה של אינרציה
for K_iter in range(2,10): #לולאה בין 2-9
    kmeans = KMeans(n_clusters=K_iter, init='random') #נאתחל את Kואת הצנטרואידים
    kmeans.fit(reduced) #נאמן את המודל
    K_list.append(K_iter) #נוסיף לרשימה
    WCSS_list.append(kmeans.inertia_) #נוסיף לרשימה את האינרציה
```
