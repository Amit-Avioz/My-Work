```python
frame["How old are you?"]
for row in frame_copy.index:
  if pd.isnull(frame_copy.loc[row,"How old are you?"]):
    frame_copy.loc[row,"How old are you?"]=frame_copy["How old are you?"].mean() #יש ערכים חסרים בעמודת גיל ולכן נעשה ממצוע של הגיל (לבדוק פקודה קצרה יותר)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

sns.catplot(x ='district',hue='Which party vote in the last election for the Knesset (March 2021):', kind ='count', data = frame,aspect=2)
```

```python
# Violinplot Displays distribution of data
# across all levels of a category.
sns.violinplot(x ='In Israeli politics it is customary to talk about left and right. Where would you place yourself on a scale where 1 is left, 4 is center and 7 is right?', y ="How old are you?", hue ="What are you?", data = frame,split=True)
```

```python
survival_stats = frame.groupby('Do you agree or disagree that Israel should have more rights to Jewish citizens than non-Jewish citizens:')['Do you agree or disagree that Israel should have more rights to Jewish citizens than non-Jewish citizens:'].count().tolist()

colors = sns.color_palette('dark')[0:5]


plt.pie(survival_stats, labels = [ "Strongly agree", "Not at all agree","Not so much agree","Pretty much agreed", "Do not know"], colors = colors, autopct='%.2f%%')
plt.show()
```

```python

group = frame.groupby(["Do you agree or disagree that Israel should have more rights to Jewish citizens than non-Jewish citizens:", '(To Jews) How do you religiously define yourself?'])
situation = group.size().unstack()

# Heatmap - Color encoded 2D representation of data.
sns.heatmap(situation, annot = True, fmt ="d")
```
